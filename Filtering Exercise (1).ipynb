# Filtering for Good: A filter for Yelp Reviews

```python
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Grab and process the raw data.
data_path = ('~/Desktop/Python Exercises/yelp_labelled.txt')
yelp_raw = pd.read_csv(data_path, delimiter= '\t', header=None)
yelp_raw.columns = ['review', 'good']
#so we can see everything to build features
pd.set_option('display.max_rows', 1000)
```

```python
yelp_raw
```

## Building Filters

```python
#selected features from the reviews that were common among reviews that were classified as 'negative.' 
#This will contribute to our specificity later.
keywords = ['disappointed','below', 'lacked', 'slow', 'terrible', 'worst', 'never', 'nasty', 'bad', 'worst', "didn't", 'nasty', 'over', 'horrible']

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    yelp_raw[str(key)] = yelp_raw.review.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
```

```python
#we're assuming that a person is more likely to use all caps when angry, as opposed to happy.
yelp_raw['allcaps'] = yelp_raw.review.str.isupper()
```

```python
#to increase sensitivity, we're also binning positive adjectives.
positive = ['love', 'great', 'best', 'enjoy', 'happy', 'amazing', 'worst', 'never', 'nasty']

for key in positive:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    yelp_raw[str(key)] = yelp_raw.review.str.contains(
        ' ' + str(key) + ' ',
        case=True
    )
```

## Correlation map

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7,5)) 
sns.heatmap(yelp_raw.corr())
#increased size of heatmap for readability
```

```python
data = yelp_raw[keywords + positive + ['allcaps']]
target = yelp_raw['good']
```

```python
data
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```

```python
print('Accuracy:') 
print((437/1000) * 100)
```

### Although accuracy is low, let's consider sensitivity and specificity.

```python
y_actual = yelp_raw.good
df_confmatrix = pd.crosstab(y_actual, y_pred)
df_confmatrix
```

```python
print('Sensitivity:')
print((df_confmatrix[1][1]/(df_confmatrix[1][1] + df_confmatrix[0][1])) *100)
```

```python
print('Specificity:')
print((df_confmatrix[1][0]/(df_confmatrix[0][0] + df_confmatrix[1][0])) *100)
```

### Since the sensitivity and specificity are high the accuracy score is far less concerning.
