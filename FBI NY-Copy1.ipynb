```python
x = df_transform.population
y = df_transform.murder_manslaughter
z = df_transform.robbery

trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    text = df_transform.city,
    mode='markers',
    marker=dict(
        size=12,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)

data = [trace1]
layout = go.Layout(
    #margin=dict(
        #l=0,
        #r=0,
        #b=0,
        #t=0
    #),
    scene = dict(
                    xaxis = dict(
                        title='Population'),
                    yaxis = dict(
                        title='Murders'),
                    zaxis = dict(
                        title='Robberies'),),
                    width=700,
                    margin=dict(
                    r=20, b=10,
                    l=10, t=10)
        )

    
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-scatter-colorscale')
```
