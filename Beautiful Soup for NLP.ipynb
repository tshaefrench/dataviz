```python
starts = list(range(10, 100, 10))
requests = 0
start = time.time()

nlp_jobs = []
for start in starts:
    my_urls = ('https://www.indeed.com/jobs?q=%22Natural+Language+Processing%22+or+%22NLP%22&fromage=last&start=' + str(start),)
    my_url = my_urls[0]
    for my_url in my_urls:
        uClient = uReq(my_url)
        html_input = uClient.read()
        uClient.close()
        soup = BeautifulSoup(html_input, "html.parser")
        for td in soup.findAll('td', {'id': 'resultsCol'}):
            part = soup.find('td', {'id': 'resultsCol'})
            a = part.findAll('a')[1]
            link = a.text.strip(), '=>', a.attrs['href']
            for a in td:
                title = soup.find("title")
                job_description = soup.find("div", {"class": "jobsearch-JobComponent-description icl-u-xs-mt--md"})
                requests += 1
                sleep(randint(1,3))
                end = time.time()
                print("Done in", end - start, "seconds")
                print('Request: {}; Frequency: {} requests/s'.format(requests, requests/end))
                clear_output(wait = True)
                posting = {
                    'title':title,
                    'job_description':job_description
                    }
                nlp_jobs.append(posting)             
                               
```

```python
print(nlp_jobs)
```

```python
with open('nlp.json', 'w') as outfile:
    json.dump(nlp_jobs, outfile)
```
