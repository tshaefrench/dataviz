```python
import numpy as np
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from pandas import DataFrame
from matplotlib import pyplot
df = pd.read_csv('~/Desktop/Python Exercises/data_titanic.csv') 
df
```

```python
pd.DataFrame({'Non-Survivors': df.groupby('Survived').get_group(0).Age,
              'Survivors':   df.groupby('Survived').get_group(1).Age}).plot.hist(stacked=True)
```

```python
pd.DataFrame({'Non-Survivors': df.groupby('Survived').get_group(0).Pclass,
              'Survivors':   df.groupby('Survived').get_group(1).Pclass}).plot.hist(stacked=True)
#This, in addition to the Fare graph below show that there are higher survival rates for those in higher passenger 
#classes and that paid more for their tickets.
```

```python
pd.DataFrame({'Non-Survivors': df.groupby('Survived').get_group(0).Fare,
              'Survivors':   df.groupby('Survived').get_group(1).Fare}).plot.hist(stacked=True)
#This shows a strong correlation between expensive fares, and rate of survival since most of the deaths that 
#occurred for those paying fares of less than $50.
```

```python
pd.DataFrame({'Non-Survivors': df.groupby('Survived').get_group(0).Age,
              'Survivors':   df.groupby('Survived').get_group(1).Age}).plot.hist(stacked=True)
#This shows that younger age promoted a higher rate of survival, as the survival rate drops off dramatically 
#after the age of 40.  This supports the notiton that children were put on life rafts sooner and 
#survived in the ocean longer afterwards.
```

```python
df = pd.read_csv('~/Desktop/Python Exercises/data_titanic.csv')
df.columns = ['Unnamed', 'PassengerID', 'Survived', 'PClass', 'Name', 'Sex', 'Age',  'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'Initial']
df
```

```python
df.Sex.replace(to_replace=dict(female=1, male=0), inplace=True)
df
```

```python
pd.DataFrame({'Non-Survivors': df.groupby('Survived').get_group(0).Sex,
              'Survivors':   df.groupby('Survived').get_group(1).Sex}).plot.hist(stacked=True)
#I converted the datatype of the Sex column, so that it could be analyzed. Men are represented by '0' and Women by '1'
#This graph shows that far more men perished on the Titanic than women. 
```
